### Image rpocessing with scikit: 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from skimage import io
from skimage.transform import rescale, resize,downscale_local_mean #reshape
```

#### Generating matrices with numpy

```python
a = np.arange(20).reshape(4,5)
print(a)
```

```python
#Ploting: 
plt.figure(figsize = (5,4))
plt.imshow(a)
plt.show()
```

#### Rescaling the image: 

```python
#Using rescale to change the scale of the matrix:
a_rescaled = rescale(a,scale=2)

print(a_rescaled.shape,a.shape)
```

```python
#PLoting
plt.figure(figsize = (5,4))
plt.imshow(a_rescaled)
plt.show()
```

```python
print(a)
print(a_rescaled)
```

```python
#Preserving the values of the oroginal values:
a_rescaled2 = rescale(a,scale=2,preserve_range = True)

print(a_rescaled2.shape,a.shape)
```

```python
print(a)
print(a_rescaled2)
```

```python
#PLoting
plt.figure(figsize = (5,4))
plt.imshow(a_rescaled2)
plt.show()
```

### Downscaling images: 

```python
a_downscaled = downscale_local_mean(a,(2,2))

print(a_downscaled)
```

```python
#PLoting
plt.figure(figsize = (5,4))
plt.imshow(a_downscaled)
plt.colorbar()
plt.show()
```
