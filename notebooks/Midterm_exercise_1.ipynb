```python
#Using usecols() to select just the hours and sea level colums
hours = pd.read_csv("/home/jdiego/Documents/Computacional_1/notebooks/h030.csv" ,sep = ",",usecols =[3])
sea_level = pd.read_csv("/home/jdiego/Documents/Computacional_1/notebooks/h030.csv" ,sep = ",",usecols =[4])

#COnverting into arrays: 
hours = np.array(hours)
sea_level = np.array(sea_level)
```
