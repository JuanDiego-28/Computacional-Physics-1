# INDEXING AS CONDITIONALS:

```python
y = np.array ([1.,2.,3.,4.,5.,0.,7.,6.,-1.])

#Printhin values higher than 0 stored in Y 
print(y>0)

#selecting the values higher than 0 and storing it in Z 
z= y[y>0]   

print(z)
```

```python
#replacling with conditionals:

z[z>5]=0

print(z)
```

# PLOTING 

```python
import matplotlib.pyplot as plt

x = np.arange(0,6*np.pi,0.01)


y= np.sin(x)

plt.figure()
plt.plot(x,y)
```

```python
plt.close()
```

```python
#The corect way of plotting

#opening the enviroment
plt.figure()    

#Plotitng
plt.plot(x,y)

#Show/save()
plt.show()
#plt.savefig()   --> DON'T FORGET TO CLOSE THE FIGURE !!! plt.close()
plt.close()
```

```python
#COSTUMIZE: 

plt.figure(figsize =(4,3))

plt.plot(x,y, color = 'red', linestyle = '-')

plt.title('SIN(X)')

plt.ylabel('sin(x)')
plt.xlabel('x')

plt.show()
plt.close()
```

# OVERPLOTING

```python
#superimpose plot: 

w=np.cos(x)
plt.figure(figsize =(4,3))

plt.plot(x,y, color = 'red', linestyle = '-')
plt.plot(x,w, color = 'green', linestyle = '-')

plt.title('SIN(X) and COS(X)')

plt.ylabel('sin(x) and cos(x)')
plt.xlabel('x')
plt.legend()
plt.show()
plt.close()
```
