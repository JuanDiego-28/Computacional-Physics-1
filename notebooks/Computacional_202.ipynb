```python
cell_image = data.cell()
```

```python
# Let's look at the image 

fig, ax = plt.subplots(figsize = (4,3))

z = ax.imshow(cell_image,cmap = 'viridis_r')

plt.colorbar(z)

plt.show()
```

```python
#Histograms: 

cell_1D = np.reshape(cell_image,(cell_image.size,1))
```

```python
plt.figure ()
plt.hist(cell_1D, histtype = 'step',bins=40)
plt.show()
```

###  Array manipulation with conditionals: 

```python
cell_1D_clean = np.where(cell_1D<100,np.nan,cell_1D)
```

```python
print(cell_1D_clean[0])
```

```python
plt.figure ()
plt.hist(cell_1D_clean, histtype = 'step',bins=40)
plt.show()
```

```python
#Reshaping into a 2D array: 

cell_clean_2D = np.reshape(cell_1D_clean,(cell_image.shape))
```

```python
print(cell_clean_2D.shape,cell_image.shape)
```

```python
fig, ax = plt.subplots(figsize = (4,3))

z = ax.imshow(cell_clean_2D,cmap = 'viridis_r')

plt.colorbar(z)

plt.show()
```

### Selecting subsets: 

```python
cell_image_subset = cell_clean_2D[300:450,350:500]
```

```python
fig, ax = plt.subplots(figsize = (4,3))

z = ax.imshow(cell_image_subset,cmap = 'viridis_r')

plt.colorbar(z)

plt.show()
```

```python
#FInding the coordinates: 
print(np.nanmax(cell_image_subset))
```

```python
#Getting indices of the maximun: 

idx_max = np.where(np.nanmax(cell_image_subset) == cell_image_subset)

print(idx_max)
```

```python
fig, ax = plt.subplots(figsize = (4,3))

z = ax.imshow(cell_image_subset,cmap = 'viridis_r')
ax.scatter(idx_max[1],idx_max[0],color='red',marker='d')
plt.colorbar(z)

plt.show()
```
