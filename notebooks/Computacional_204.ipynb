### BINARI FILES

```python
#Importing the file:
beta_temp = open('/home/jdiego/Documents/Computacional_1/simulation_data/beta_temp.dbl', 'rb') #rb = binari file
```

```python
print(beta_temp)
```

```python
#Turn the dile into readable object: 

data_beta = np.fromfile(beta_temp, dtype= np.float64)

print(type(data_beta))

print(data_beta.shape)
```

```python
# Reshape into a 2D array
data_2D = data_beta.reshape((160,160))
# Plotting 

plt.figure()

plt.imshow(np.log10(data_2D), cmap = "Greens")

plt.colorbar()
plt.show()
```
