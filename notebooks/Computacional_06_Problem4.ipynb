```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

#### Creating the reading funtion:

```python
def read_data (file_name):
    '''
    Input: Data file
    Output: Retunrs 4 arrays that are: Time,mixedmass 1,2 and 3 
    '''
    
    df= pd.read_csv(file_name, sep= '\s+')
    
    #getting the keys of the file:
    keys = df.keys()
    
    #converting to arrays: 
    time = np.array(df[keys[0]])
    mix_1 = np.array(df[keys[1]])
    mix_2 = np.array(df[keys[2]])
    mix_3 = np.array(df[keys[3]])
    

    return time, mix_1 ,mix_2 ,mix_3 
```

```python
file_name = '/home/jdiego/Downloads/data_4.txt'

time, mix_1 ,mix_2 ,mix_3 = read_data(file_name)

#Proving the function works: 
#print(time, mix_1 ,mix_2 ,mix_3 )
```

#### Plotting the vectors:

```python
#Doing the plot of the mix: 

plt.figure()
plt.plot(time,mix_1, color = 'b' , label= "mix_1",linestyle= '-')
plt.plot(time,mix_2, color = 'g',label= "mix_2",linestyle= '--')
plt.plot(time,mix_3,color= 'r',label= "mix_3",linestyle= ':')

#calling the legends:
plt.legend()

#lAbeling:
plt.xlabel('time(s)')
plt.ylabel('mixtures(g)')

#Showing the figure
plt.show()
```

#### Creating the fraction function:

```python
def mixing_franctions(mix_i):
    '''
    INPUT: recieves a vector that saves the values of the mix
    OUTPUT: returns a vector divided by the first objecto of the vector
    '''
    m0= 61.18026
    
    mixing_frac =  mix_i/m0
    
    return (mixing_frac)
```

```python
mix_frac_1=mixing_franctions(mix_1)
mix_frac_2=mixing_franctions(mix_2)
mix_frac_3=mixing_franctions(mix_3)
```

####  Plotting the mixed fractions:

```python
#Doing the plot of the mix: 

plt.figure()
plt.plot(time,mix_frac_1, color = 'b' , label= "mix_frac_1",linestyle= '-')
plt.plot(time,mix_frac_2, color = 'g',label= "mix_frac_2",linestyle= '--')
plt.plot(time,mix_frac_3,color= 'r',label= "mix_frac_3",linestyle= ':')

#calling the legends:
plt.legend()

#lAbeling:
plt.xlabel('time(s)')
plt.ylabel('mixtures(g)')

#Showing the figure
plt.show()
```

#### Creating the total_mass function:

```python
def total_mix(mix_1,mix_2,mix_3):
    '''
    INPUT: Recives the three vectors of the mix
    OUTPUT: retunrs the sum of the three vectors
    '''
    total = (mix_1 + mix_2 + mix_3)
    
    return total


total = total_mix(mix_1,mix_2,mix_3)

#Checking the function works:
#print(total)
```

####  Creating the total_mix_frac function:

```python
def total_frac_mix(total):
    '''
    INPUT: Recieves the vector containgn the total mix
    OUTPUT: Retunrs a vector divided by m0
    '''
    m0= 61.18026 
    total_frac = total/m0
    
    return total_frac  

total_frac = total_frac_mix(total)
#checking it works:
#print(total_frac)
```
